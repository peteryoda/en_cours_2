```python
def strate_percentiles(x):
    
    if x > 0 and x <= perc_values[0]:
        resultat = '1'
#         print (resultat)
    elif x > perc_values[0] and x <= perc_values[1]:
        resultat = '1'
#         print (resultat)
    elif x > perc_values[1] and x <= perc_values[2]:
        resultat = '2'
#         print (resultat)
    elif x > perc_values[2] and x <= perc_values[3]:
        resultat = '3'
#         print (resultat)
    elif x > perc_values[3] and x <= perc_values[4]:
        resultat = '4' 
#         print (resultat)
    elif x > perc_values[4] and x <= perc_values[5]:
        resultat = '5'
#         print (resultat)
    elif x > perc_values[5] and x <= perc_values[6]:
        resultat = '6'
#         print (resultat)
    elif x > perc_values[6] and x <= perc_values[7]:
        resultat = '7'
#         print (resultat)
    elif x > perc_values[7] and x <= perc_values[8]:
        resultat = '8'
#         print (resultat)
    elif x > perc_values[8] and x <= perc_values[9]:
        resultat = '9'
#         print (resultat)
#     elif x > perc_values[9] and x <= 1:
    else :
        resultat = '10'
#         print (resultat)

    return resultat

all_products_matched['strate_var_score'] = all_products_matched['confidence_score'].map(strate_percentiles)
```

```python
df_with_url_crf_and_ntr = all_sampled_couples[(all_sampled_couples['distributeur_one'] == 'crf') |
                    (all_sampled_couples['distributeur_one'] == 'ntr') |
                    (all_sampled_couples['distributeur_two'] == 'crf') |
                    (all_sampled_couples['distributeur_two'] == 'ntr')]
```

```python
liste_urls = []

for row in df_with_url_crf_and_ntr.itertuples():
    if row.distributeur_one == 'crf' or row.distributeur_one == 'ntr':
        if row.URL_one not in liste_urls:
            liste_urls.append(row.URL_one)
    if row.distributeur_two == 'crf' or row.distributeur_two == 'ntr':
        if row.URL_two not in liste_urls:
            liste_urls.append(row.URL_two)
```
